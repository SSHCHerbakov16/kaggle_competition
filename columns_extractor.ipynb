```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data = pd.read_csv('train.csv')
#data = data.drop('Target',axis=1)
data
```

```python
fin_test_data = pd.read_csv('test.csv')
test_data = fin_test_data.drop('index',axis=1)
test_data
```

```python
x = data.drop('Target',axis=1)
y = data.Target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,50),max_iter=500)
mlp.fit(x_train,y_train)
```

```python
predictions = mlp.predict(x_test)
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
fin_test_data = pd.read_csv('test.csv')
#test_data = fin_test_data.drop('id',axis=1)
test_data = scaler.fit_transform(test_data)
```

```python
fin_predictions = mlp.predict(test_data)
fin_predictions
```

```python
output = pd.DataFrame({'index':fin_test_data.index+1,'Target':fin_predictions})
output.to_csv('answer.csv', index=False)
print("ok")
```
